```python
def data_discovery(file):#фунция для других Вендорлог для исследования dau
    import pandas as pd
    from matplotlib import pyplot as plt
    pd = pd.read_excel(f'{file}.xlsx')
    dau = pd.pivot_table(index = 'Дата сварки', columns = 'Тип стыка2', values = '№ Стыка', aggfunc = 'count', fill_value=0)
    print(dau.head(3))
                      
    
    dau.plot(figsize = (15, 5), grid = True) 
    plt.title('Изменение DAU с разбивкой по типу стыка'), plt.xlabel('Дата'), plt.ylabel('Тип стыка');
                      

    dau['per_C17'] = dau['C17'].sum() / dau.sum(axis = 1)
    dau['per_C8'] = dau['C8'].sum() / dau.sum(axis = 1)
    dau['per_Y17'] = dau['У17'].sum() / dau.sum(axis = 1)
                      
    dau[['per_C17', 'per_C8', 'per_Y17']].plot.area(figsize = (15, 15))
    plt.title('Процентное изменение DAU по типу стыка'), plt.xlabel('Дата'), plt.ylabel('Стыки');
    return data_discovery
    
```

```python
x = data_discovery('вендорлог YMT-1-110-HP-412002-01B-16-VT-22')
```

```python
x.to_excel('dau_func.xlsx', index = False)
```
