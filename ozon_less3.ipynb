```python
new_lesson3['Дата (без времени)'] = new_lesson3['Дата сварки'].dt.date
new_lesson3['Год-Месяц'] = new_lesson3['Дата сварки'].dt.to_period('M')
new_lesson3.head(10)
```

```python
dau = new_lesson3.groupby(['Дата (без времени)']).agg({'№ Стыка': 'nunique'}) #количество уникальных сваренных стыков за определенную дату
dau.columns = ['DAU']
dau.head(10)
```

```python
from matplotlib import pyplot as plt

dau.plot(figsize = (15, 5), grid = True, color = 'magenta', linestyle = ':')
plt.title('Динамика DAU'), plt.xlabel('Дата'), plt.ylabel('Уникальные сваренные стыки')

plt.show()
```

```python
dau = new_lesson3.pivot_table(index = 'Дата сварки', columns = 'Тип стыка2', values = '№ Стыка', aggfunc = 'count', fill_value=0) #стыки, разбитые по дате и типу

dau.head(5)
```

```python
from datetime import datetime

ax = dau.plot(figsize = (15, 5), grid = True)
ax.axvline(datetime(2020, 2, 1), color = 'grey') #дата смены сварщика
plt.title('Изменение DAU с разбивкой по типу стыка'), plt.xlabel('Дата'), plt.ylabel('Тип стыка');
```

```python
dau.head(10)
```

```python
dau['Общий_dau'] = dau.sum(axis = 1)
dau.head(10
        )
```

```python
dau['Процент С8'] = dau['C8'] / dau['Общий_dau']
dau.head(5).style.format({'Процент С8': '{:.3%}'})
```

```python
def get_perc_except_C8(cell_value):
    ''''Функция вычисляет общий процент стыков, без типа С8 '''
    return 1 - cell_value

dau['Проценты без C8'] = dau['Процент С8'].apply(get_perc_except_C8)
dau.head(5)
```
